# Expected value of empirical risk for OLS 

### Question 7

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100  # Number of samples
d = 10   # Number of features
theoretical_sigma = 2  # Theoretical sigma value

np.random.seed(42)
X = np.random.randn(n, d)
theta_star = np.random.randn(d)
epsilon = np.random.randn(n) * theoretical_sigma

# Response variable y
y = X @ theta_star + epsilon

# OLS estimator
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Residuals
residuals = y - X @ theta

# Sigma^2 using the result of Question 6
sigma_square = np.sum(residuals**2) / (n - np.trace(X @ np.linalg.inv(X.T @ X) @ X.T))

print(f"True sigma^2: {theoretical_sigma**2}")
print(f"Estimated sigma^2 (from simulation): {sigma_square}")

# Residuals
plt.hist(residuals, bins=30, alpha=0.5, label='Residuals')
plt.axvline(x=0, color='red', linestyle='--')
plt.title('Histogram of Residuals')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

We can see that the result is consistent with the theoretical value we have chosen (2^2 => 4) as it is close to the estimated sigma square (4.27)
